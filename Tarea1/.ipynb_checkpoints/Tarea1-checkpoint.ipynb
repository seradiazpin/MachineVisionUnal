### Interpolacion vecino más cercano

```python
%matplotlib inline
import cv2
from IPython.display import Image
import numpy as np
from matplotlib import pyplot as plt
from __future__ import division

img = cv2.imread('Fig/493-tierra.png',0)

def resize(image,d,inter):
    width, height = img.shape[:2]
    print("size", width*height)
    img_resizeN = np.zeros((width*d,height*d))
    
    widthRz, heightRz = img_resizeN.shape[:2]
    
    print width ,height
    x_ratio = width/widthRz
    y_ratio = height/heightRz
    
    x_ratioBL = (width-1)/widthRz
    y_ratioBL = (height-1)/heightRz
    print x_ratio ,y_ratio
    px, py  = 0 , 0
    for i in range(widthRz-1):
        for j in range(heightRz-1):
            if(inter == "N"):
                img_resizeN[i][j] = nearest(image,i,j,x_ratio,y_ratio);
    return img_resizeN
print "Listo"
```

```python
#-nearest
def nearest(img,x,y,x_ratio,y_ratio):
    px = np.floor(x*x_ratio)
    py = np.floor(y*y_ratio)
    return img[int(px)][int(py)]

reziceN_img = resize(img,5,"N")
plt.figure(1)
plt.subplot(221), plt.imshow(img, cmap=plt.cm.gray), plt.title('Imagen')
plt.subplot(222), plt.imshow(reziceN_img, cmap=plt.cm.gray,interpolation='none',aspect='auto'), plt.title('Nearest')
```

```python
def bilinear(img,i,j,x_ratioBL,y_ratioBL):
    #Bilinear
    x = int(i*x_ratioBL)
    y = int(j*y_ratioBL)
    x_diff = (x_ratioBL * i) - x ;
    y_diff = (y_ratioBL * j) - y ;
    #puntos para interpolar
    #A = img_array[index]
    A = img[x][y]
    #B = img_array[index+1]
    B = img[x+1][y]
    #C = img_array[index+i]
    C = img[x][y+1]
    #D = img_array[index+i+1]
    D = img[x+1][y+1]

    V = int(A*(1-x_diff)*(1-y_diff) +  B*(x_diff)*(1-y_diff) + C*(y_diff)*(1-x_diff)+  D*(x_diff*y_diff))
    return V
```

```python





def resize(image):
    width, height = img.shape[:2]
    img_array = np.asarray(image).reshape(-1)
    print("size",len(img_array))
    img_resizeN = np.zeros((width*10,height*10))
    img_resizeBL = np.zeros((width*10,height*10))
    img_resizeBC = np.zeros((width*10,height*10))
    """
    for i in range(width):
        for j in range(height):
            di = i * 5
            dj = j * 5
            img_resize[di][dj] = image[i][j]
    
    """
    widthRz, heightRz = img_resizeN.shape[:2]
    print width ,height
    x_ratio = width/widthRz
    y_ratio = height/heightRz
    
    x_ratioBL = (width-1)/widthRz
    y_ratioBL = (height-1)/heightRz
    print x_ratio ,y_ratio
    px, py  = 0 , 0
    for i in range(widthRz-1):
        for j in range(heightRz-1):
            #Nearest
            px = np.floor(i*x_ratio)
            py = np.floor(j*y_ratio)
            img_resizeN[i][j] = img[int(px)][int(py)] ;
            
            #Bilinear
            x = int(i*x_ratioBL)
            y = int(j*y_ratioBL)
            x_diff = (x_ratioBL * i) - x ;
            y_diff = (y_ratioBL * j) - y ;
            
            index = x*i+y

            #puntos para interpolar

            #A = img_array[index]
            A = img[x][y]
            #B = img_array[index+1]
            B = img[x+1][y]
            #C = img_array[index+i]
            C = img[x][y+1]
            #D = img_array[index+i+1]
            D = img[x+1][y+1]
            
            
            V = int(A*(1-x_diff)*(1-y_diff) +  B*(x_diff)*(1-y_diff) + C*(y_diff)*(1-x_diff)+  D*(x_diff*y_diff))
            img_resizeBL[i][j] = V
            #Bicubic
    bufferPixels = np.zeros((4,4))
    
    x_ratioBC = (width-3)/widthRz
    y_ratioBC = (height-3)/heightRz
    
    
    
    for i in range(widthRz-3):
        for j in range(heightRz-3):
            
            xO = i*x_ratioBC
            yO = j*y_ratioBC
            
            x = int(i*x_ratioBC)
            y = int(j*y_ratioBC)
            
            xfract = xO - np.floor(x);
            yfract = yO - np.floor(y);
            
            bufferPixels[0,:] = img[x:x+4,y]
            bufferPixels[1,:] = img[x:x+4,y+1]
            bufferPixels[2,:] = img[x:x+4,y+2]
            bufferPixels[3,:] = img[x:x+4,y+3]
            
            img_resizeBC[i][j] = int(clamp(bicubic(bufferPixels,xfract,yfract),0,255))
    
    return img_resizeN , img_resizeBL, img_resizeBC

def cubic(p, t):

    a = -p[0] / 2.0 + (3.0*p[1]) / 2.0 - (3.0*p[2]) / 2.0 + p[3] / 2.0;
    b = p[0] - (5.0*p[1]) / 2.0 + 2.0*p[2] - p[3] / 2.0
    c = -p[0] / 2.0 + p[2] / 2.0
    d = p[1]
    return a*t*t*t + b*t*t + c*t + d

def clamp(v, minv, maxv):
    if (v < minv):
        return minv
    else:
        if (v > maxv):
            return maxv
    return v

def bicubic(pixels,x,y):
    arr = np.zeros((4,1))
    arr[0] = cubic(pixels[0], y)
    arr[1] = cubic(pixels[1], y)
    arr[2] = cubic(pixels[2], y)
    arr[3] = cubic(pixels[3], y)
    return cubic(arr, x)

reziceN_img, reziceBL_img ,reziceBC_img= resize(img)


plt.figure(1)
plt.subplot(221), plt.imshow(img, cmap=plt.cm.gray), plt.title('Imagen')
plt.subplot(222), plt.imshow(reziceN_img, cmap=plt.cm.gray,interpolation='none',aspect='auto'), plt.title('Nearest')
plt.subplot(223), plt.imshow(reziceBL_img, cmap=plt.cm.gray,interpolation='none',aspect='auto'), plt.title('Bilinear')

plt.subplot(224), plt.imshow(reziceBC_img, cmap=plt.cm.gray,interpolation='none',aspect='auto'), plt.title('Bicubic')


#interpol = vecinoMasCercanos([[2,2],[3,1],[4,9],[5,1]],0.1)
#plt.plot(*zip(*interpol)), plt.title('Histograma')
```
